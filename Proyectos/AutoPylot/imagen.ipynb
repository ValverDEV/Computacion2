```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from Regresiones import NumMethods
from sympy import symbols
x = symbols('x')
```

```python
n = len(img)

Xs = []
Ys = []

matriz = np.zeros((n,n))
for row in range(99,-1,-1):
    for col in range(99,-1,-1):
        if img[row][col][0] > 0:
            matriz[row][col] = 1
            print(row, col)
            Xs.append(100-row)
            Ys.append(100-col)
```

```python
plt.scatter(Xs,Ys)
```

```python
def split_lines(Xs, Ys):
    dist = 20
    Xs1 = []
    Ys1 = []
    
    Xs2 = [Xs[0]]
    Ys2 = [Ys[0]]

    print(Xs2, Ys2)

    for i in range(1, len(Xs)):
        if not(abs(Xs2[-1] - Xs[i]) <= dist and abs(Ys2[-1] - Ys[i]) <= dist):
            Xs1.append(Xs[i])
            Ys1.append(Ys[i])
            break

    for i in range(1, len(Xs)):
        # print(i)
        if abs(Xs2[-1] - Xs[i]) <= dist and abs(Ys2[-1] - Ys[i]) <= dist:
            # print('estamos en el 2o')
            # print(f'anteriores {Xs2[-1]}, {Ys2[-1]}, se añaden {Xs[i], Ys[i]}')
            Xs2.append(Xs[i])
            Ys2.append(Ys[i])
        else:
        # elif abs(Xs1[-1] - Xs[i]) <= 3 and abs(Ys1[-1] - Ys[i]) <= 3:
            # print('estmaos en el 1o')
        #     if Xs1:
            # print(f'anteriores {Xs1[-1]}, {Ys1[-1]}, se añaden {Xs[i], Ys[i]}')
            Xs1.append(Xs[i])
            Ys1.append(Ys[i])

    return Xs1, Ys1, Xs2, Ys2
```

```python
X1, Y1, X2, Y2 = split_lines(Xs, Ys)
```

```python
plt.scatter(X1,Y1, color='g')
plt.scatter(X2,Y2, color='r')
plt.show()
```

```python
NM = NumMethods(X1, Y1)
lin = NM.linear()
NM.grafica()
poli1 = NM.polinomial(2)
NM.grafica()
```

```python
NM = NumMethods(X2, Y2)
lin = NM.linear()
NM.grafica()
poli2 = NM.polinomial(2)
NM.grafica()
```

```python
print(lin)
print(lin.diff(x))
```

```python
a = np.arange(0,100, 1)
b1 = [poli1.subs(x, i) for i in a]
b2 = [poli2.subs(x, i) for i in a]
```

```python
plt.plot(a, b1)
plt.plot(a, b2)
plt.scatter(Xs, Ys)
plt.show()
```

```python
final = (poli2 + poli1)/2
b3 = [final.subs(x, i) for i in a]
```

```python
plt.plot(a, b1)
plt.plot(a, b2)
plt.plot(a, b3)
plt.scatter(Xs, Ys)
plt.show()
```
