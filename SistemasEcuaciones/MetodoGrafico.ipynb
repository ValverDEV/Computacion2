```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
```

```python
x, y, z = symbols('x y z')
```

```python
# expr = sympify(input('Ingese la ecuación:\n'))
expr = x**3 + 5*x**2 -10
expr
```

```python
a = np.arange(-10,10,0.001)
```

```python
f = lambdify(x, expr, 'numpy')
b = f(a)
```

```python
raizX = []
raizY = []
for i in range(0,len(b)):
    print(i)
    print(b[i])
    if round(b[i], 1) == 0:
        raizX.append(a[i])
        raizY.append(b[i])
```

```python
print(raizX)
```

```python
plt.plot(a,b)
plt.scatter(raizX, raizY)
```
